```python
import plotly.express as px  # pip install plotly
# pip install dash-bootstrap-components
import dash_bootstrap_components as dbc
from dash_extensions import Lottie
import plotly.graph_objects as go
from dash.dependencies import Output, Input
import dash_core_components as dcc
import dash_html_components as html
from jupyter_dash import JupyterDash  # pip install dash
import pandas as pd
import warnings
warnings.filterwarnings('ignore')

url_position = "https://assets7.lottiefiles.com/packages/lf20_vqtpkrb8.json"
url_wins = "https://assets6.lottiefiles.com/packages/lf20_XKqN79.json"
url_draws = "https://assets5.lottiefiles.com/packages/lf20_ee2kjf6e.json"
url_losses = "https://assets9.lottiefiles.com/packages/lf20_jz6xkank.json"
url_goals = "https://assets6.lottiefiles.com/packages/lf20_3SQyhp.json"
url_points = "https://assets5.lottiefiles.com/datafiles/BNDkuotsD4cSeYx/data.json"

options = dict(loop=True, autoplay=True, rendererSettings=dict(
    preserveAspectRatio='xMidYMid slice'))

df = pd.read_csv('EPL Standings 2010-2021.csv')

app = JupyterDash(__name__, external_stylesheets=[dbc.themes.LUX])

app.layout = dbc.Container([
    html.Br(),
    html.H1("Premier League Stats 2010-2020", style={'text-align': 'center'}),
    html.Br(),
    dbc.Row([
        dbc.Col([
            dbc.Card([
                html.Img(src='/assets/pexels-photo-3991976.jpeg',
                         style={'width': '1300px',
                               'height': '450px'})
                #                 dbc.CardImg(src='/assets/Premier_League_Logo.svg.png', top=True)
            ], className='mb-5'),
        ]),

    ]),
    html.Br(),
    html.H2("Team Statistics", style={'text-align': 'left'}),
    html.Br(),
    dbc.Row([
        #         html.H3("Stats for: ", style={'text-align': 'left'}),
        dcc.Dropdown(id="team_year_stats",
                     options=[
                         {"label": "2010", "value": '2010-11'},
                         {"label": "2011", "value": '2011-12'},
                         {"label": "2012", "value": '2012-13'},
                         {"label": "2013", "value": '2013-14'},
                         {"label": "2014", "value": '2014-15'},
                         {"label": "2015", "value": '2015-16'},
                         {"label": "2016", "value": '2016-17'},
                         {"label": "2017", "value": '2017-18'},
                         {"label": "2018", "value": '2018-19'},
                         {"label": "2019", "value": '2019-20'},
                         {"label": "2020", "value": '2020-21'}],
                     multi=False,
                     value='2010-11',
                     style={'width': "40%"}
                     ),

        dcc.Dropdown(id="seected_team",
                     options=[
                        {"label": "Chelsea", "value": 'Chelsea'},
                        {"label": "Tottenham Hotspur",
                            "value": 'Tottenham Hotspur'},
                        {"label": "Manchester City", "value": 'Manchester City'},
                        {"label": "EvertonEverton", "value": 'Everton'},
                        {"label": "Manchester United",
                            "value": 'Manchester United'},
                        {"label": "Arsenal", "value": 'Arsenal'},
                        {"label": "Liverpool", "value": 'Liverpool'},
                        {"label": "West Ham United", "value": 'West Ham United'},

                     ],
                     multi=False,
                     value='Chelsea',
                     style={'width': "40%"}
                     ),

    ]),

    html.Br(),
    
   
    
    html.Br(),
    dbc.Row([
        dbc.Col([
            dbc.Card([
                dbc.CardHeader(Lottie(options=options, width="67%",
                                      height="67%", url=url_position)),
                dbc.CardBody([
                    html.H6('Position'),
                    html.H2(id='content-position', children="000")
                ], style={'textAlign': 'center'})
            ]),
        ], width=2),
        dbc.Col([
            dbc.Card([
                dbc.CardHeader(
                    Lottie(options=options, width="67%", height="67%", url=url_wins)),
                dbc.CardBody([
                    html.H6('Wins'),
                    html.H2(id='content-wins', children="000")
                ], style={'textAlign': 'center'})
            ]),
        ], width=2),
        dbc.Col([
            dbc.Card([
                dbc.CardHeader(
                    Lottie(options=options, width="80%", height="80%", url=url_draws)),
                dbc.CardBody([
                    html.H6('Draws'),
                    html.H2(id='content-draws', children="000")
                ], style={'textAlign': 'center'})
            ]),
        ], width=2),
        dbc.Col([
            dbc.Card([
                dbc.CardHeader(
                    Lottie(options=options, width="67%", height="67%", url=url_losses)),
                dbc.CardBody([
                    html.H6('Losses'),
                    html.H2(id='content-losses', children="000")
                ], style={'textAlign': 'center'})
            ]),
        ], width=2),
        dbc.Col([
            dbc.Card([
                dbc.CardHeader(
                    Lottie(options=options, width="67%", height="67%", url=url_goals)),
                dbc.CardBody([
                    html.H6('Goals'),
                    html.H2(id='content-goals', children="000")
                ], style={'textAlign': 'center'})
            ]),
        ], width=2),
        dbc.Col([
            dbc.Card([
                dbc.CardHeader(
                    Lottie(options=options, width="67%", height="67%", url=url_points)),
                dbc.CardBody([
                    html.H6('Points'),
                    html.H2(id='content-points', children="000")
                ], style={'textAlign': 'center'})
            ]),
        ], width=2),
    ], className='mb-2'),

    html.Br(),
    dcc.Dropdown(id="slct_year",
                 options=[
                     {"label": "2010", "value": '2010-11'},
                     {"label": "2011", "value": '2011-12'},
                     {"label": "2012", "value": '2012-13'},
                     {"label": "2013", "value": '2013-14'},
                     {"label": "2014", "value": '2014-15'},
                     {"label": "2015", "value": '2015-16'},
                     {"label": "2016", "value": '2016-17'},
                     {"label": "2017", "value": '2017-18'},
                     {"label": "2018", "value": '2018-19'},
                     {"label": "2019", "value": '2019-20'},
                     {"label": "2020", "value": '2020-21'}],
                 multi=False,
                 value='2010-11',
                 style={'width': "40%"}
                 ),
    html.Br(),
    dbc.Row([

        dbc.Col([
            dbc.Card([
                dbc.CardBody([
                    dcc.Graph(id='first-chart', figure={}, )
                ])
            ]),
        ], width=6),

        dbc.Col([
            dbc.Card([
                dbc.CardBody([
                    dcc.Graph(id='second-chart', figure={}, )
                ])
            ]),
        ], width=6),

    ]),

    html.Br(),
    
    html.H4("qualified and Relegated Team", style={'text-align': 'left'}),
    
    html.Br(),
    
    dbc.Row([
        dcc.Dropdown(id="second_slct_year",
                     options=[
                         {"label": "2010", "value": '2010-11'},
                         {"label": "2011", "value": '2011-12'},
                         {"label": "2012", "value": '2012-13'},
                         {"label": "2013", "value": '2013-14'},
                         {"label": "2014", "value": '2014-15'},
                         {"label": "2015", "value": '2015-16'},
                         {"label": "2016", "value": '2016-17'},
                         {"label": "2017", "value": '2017-18'},
                         {"label": "2018", "value": '2018-19'},
                         {"label": "2019", "value": '2019-20'},
                         {"label": "2020", "value": '2020-21'}],
                     multi=False,
                     value='2010-11',
                     style={'width': "40%"}
                     ),

        dcc.Dropdown(id="qualification",
                     options=[
                         {"label": "Qualification", "value": 'Qualificat'},
                         {"label": "Relegation", "value": 'Relegation'}, ],
                     multi=False,
                     value='Qualificat',
                     style={'width': "40%"}
                     ),
    ]),

    html.Br(),
    html.Br(),

    dbc.Col([
        dbc.Card([
            dbc.CardBody([
                dcc.Graph(id='third-chart', figure={}, )
            ])
        ]),
    ], width=12),
    
    

    html.Br(),
    
    html.H4("Season stats", style={'text-align': 'left'}),
    
    html.Br(),

    dbc.Row([
        dcc.Dropdown(id="third_slct_year",
                     options=[
                         {"label": "2010", "value": '2010-11'},
                         {"label": "2011", "value": '2011-12'},
                         {"label": "2012", "value": '2012-13'},
                         {"label": "2013", "value": '2013-14'},
                         {"label": "2014", "value": '2014-15'},
                         {"label": "2015", "value": '2015-16'},
                         {"label": "2016", "value": '2016-17'},
                         {"label": "2017", "value": '2017-18'},
                         {"label": "2018", "value": '2018-19'},
                         {"label": "2019", "value": '2019-20'},
                         {"label": "2020", "value": '2020-21'}],
                     multi=False,
                     value='2010-11',
                     style={'width': "40%"}
                     ),

        dcc.Dropdown(id="select_column",
                     options=[
                         {"label": "Win", "value": 'W'},
                         {"label": "Loss", "value": 'L'},
                         {"label": "Draw", "value": 'D'},
                         {"label": "Points", "value": 'Pts'},
                         {"label": "GF", "value": 'GF'},
                         {"label": "GA", "value": 'GA'},
                         {"label": "GD", "value": 'GD'}],
                     multi=False,
                     value='W',
                     style={'width': "40%"}
                     ),
    ]),
    html.Br(),
    html.Br(),

    dbc.Col([
        dbc.Card([
            dbc.CardBody([
                dcc.Graph(id='forth-chart', figure={}, )
            ])
        ]),
    ], width=12),

    #     'Pos', 'Team', 'Pld', 'W', 'D', 'L', 'GF', 'GA', 'GD', 'Pts',

])


@app.callback(
    Output('content-position', 'children'),
    Output('content-wins', 'children'),
    Output('content-draws', 'children'),
    Output('content-losses', 'children'),
    Output('content-goals', 'children'),
    Output('content-points', 'children'),
    Input('team_year_stats', 'value'),
    Input('seected_team', 'value'),
)
def team_stats_for_specific_year(selected_year, selected_team):
    record = df[(df['Team'] == selected_team) &
                (df['Season'] == selected_year)]
    return record['Pos'], record['W'], record['D'], record['L'], record['GF'], record['Pts']


@app.callback(
    Output('first-chart', 'figure'),
    Input('slct_year', 'value'),
)
def season_stats(slct_year):
    print(slct_year)
    df_season = df[df['Season'] == slct_year]
    fig_line = px.histogram(df_season, x="Team", y='Pts',
                            labels={
                                "Team": "Teams"
                            },
                            title=f"Premier League Season {slct_year}")
    return fig_line


@app.callback(
    Output('second-chart', 'figure'),
    Input('slct_year', 'value'),
)
def seasons_winners(temp):
    df_winers = df[df['Pos'] == 1]
    count = df_winers.Team.value_counts()
    values = count.to_dict()

    fig = px.pie(df_winers, names=list(values.keys()), values=list(values.values()),
                 title='Winners From 2010 to 2020'
                 )
    fig.update_traces(hoverinfo='label+percent', textinfo='label+value', )
    return fig


@app.callback(
    Output('third-chart', 'figure'),
    Input('second_slct_year', 'value'),
    Input('qualification', 'value'),
)
def qualification_relegation(year, qualification):
    season = df[(df['Season'] == year) &
                (df['Qualification or relegation'].str[:10] == qualification)][['Team', 'Qualification or relegation']]

    fig = go.Figure(data=[go.Table(header=dict(values=['Team', 'Status']),
                                   cells=dict(values=[list(season.Team), list(season['Qualification or relegation'])]))
                          ])

    return fig


@app.callback(
    Output('forth-chart', 'figure'),
    Input('third_slct_year', 'value'),
    Input('select_column', 'value'),
)
def select_column_stats(year, column):
    fig = px.histogram(df[df['Season'] == year].sort_values(
        by=column, ascending=False), y=column, x='Team')
    return fig


app.run_server(mode='external', port=8012)

# if __name__ == '__main__':
#     app.run_server(debug=True)
```
